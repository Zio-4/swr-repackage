```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('ldt-design-item-difficulties.csv')
```

```python
df = df.sort_values(["block", "realpseudo","b_i"], ascending = (True,True,True))
```

```python
df = df.reset_index()
```

```python
df = df.drop(['index','Unnamed: 0'],axis = 1)
```

```python
df
```

```python
def label_difficulty(local_df,num_levels,num_blocks):
    n = len(local_df)
    lis = []
    for i in range(2*num_blocks):
        for j in range(num_levels):
            for k in range(int(n/(num_blocks * num_levels * 2))):
                    lis.append(j)
    print(lis)
    return local_df.insert(6,"difficulty_group",lis)
```

```python
label_difficulty(df,6,3)
```
